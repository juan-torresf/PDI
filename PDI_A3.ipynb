**Actividad 3 Retroalimentación del análisis y desarrollo del histograma**

```python
#importar librerías
import cv2,urllib.request,matplotlib.pyplot as plt
```

```python
#descargar y cargar la imagen
url = 'https://raw.githubusercontent.com/dilp90/PDI/main/Imagenes/ave.jpg'
urllib.request.urlretrieve(url,'ave.jpg')
imagen = cv2.imread('ave.jpg')

imagen_rgb = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(imagen_rgb)
plt.title('imagen original')
plt.axis('off')
plt.show()
```

```python
#convertir la imagen a escala de grises
gris = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(6,6))
plt.imshow(gris,cmap='gray')
plt.title('imagen en escala de grises')
plt.axis('off')
plt.show()
```

```python
#calcular el histograma
histograma = cv2.calcHist([gris],[0],None,[256],[0,256])

plt.figure(figsize=(8,4))
plt.plot(histograma,color='black')
plt.title('histograma con cv2.calcHist()')
plt.xlabel('nivel de intensidad')
plt.ylabel('número de píxeles')
plt.grid(True)
plt.show()
```

```python
#calcular el histograma con matplotlib
plt.figure(figsize=(8,4))
plt.hist(gris.ravel(),bins=256,range=[0,256],color='gray')
plt.title('histograma con plt.hist()')
plt.xlabel('nivel de intensidad')
plt.ylabel('número de píxeles')
plt.grid(True)
plt.show()
```

```python
#visualizar la imagen y su histograma
fig,ax = plt.subplots(1,2,figsize=(12,5))

ax[0].imshow(gris,cmap='gray')
ax[0].set_title('imagen en escala de grises')
ax[0].axis('off')

ax[1].plot(histograma,color='black')
ax[1].set_title('histograma')
ax[1].set_xlabel('nivel de intensidad')
ax[1].set_ylabel('número de píxeles')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
